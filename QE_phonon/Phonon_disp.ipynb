## Phonon dispersions

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt("./NaNNO.freq.gp")
plt.figure(figsize=(12, 8), dpi=400)
nbands = data.shape[1] - 1
for band in range(nbands):
    plt.plot(data[:, 0], data[:, band], linewidth=1, alpha=0.5, color='k')
    #plt.scatter(data[:, 0], data[:, band], marker='.', alpha=0.5, color='k')
# High symmetry k-points (check matdyn.GaAs.in)
plt.axvline(x=data[0, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[50, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[100, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[150, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[200, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[250, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[300, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[350, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[400, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[450, 0], linewidth=0.5, color='k', alpha=0.5)
plt.axvline(x=data[500, 0], linewidth=0.5, color='k', alpha=0.5)
plt.xticks(ticks= [0, data[50, 0], data[100, 0], data[150, 0],data[200, 0],data[250, 0],data[300, 0], data[350, 0],data[400, 0],data[450, 0],data[500, 0], data[-1, 0]], \
           labels=['Z','R','A','Z','$\Gamma$','M', 'X', '$\Gamma$', 'A','M','R','X'])
plt.ylabel("Frequency (cm$^{-1}$)")
plt.xlim(data[0, 0], data[-1, 0])
plt.ylim(-200 ,500 )
plt.title('NaNNO phonon dispersion DFPT')
plt.show()
```
